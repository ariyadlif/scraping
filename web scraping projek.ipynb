```python
#masukan dalam 1 file bahan yang sudah di install tadi
#seperti selenium, BeautifulSoup kemudian import
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#chromedriver file yang sudah di download dan perintah ini 
#akan membuka chrome baru secara otomatis
driver = webdriver.Chrome("chromedriver")
```

```python
#ini membuat variabel dengan data array kosong
#yang nantinya akan di gunakan untuk excel
desk=[]
prices=[]
image=[]
namaprod=[]
#ini link url web yang akan digunakan untuk scaping nya
driver.get("https://www.ikea.co.id/in/search/?q=meja")
```

```python
#content yaitu variabel
#driver.page_source ini koneksi link url web scraping na 
#variabel soup isinya variabel data content dan diformat kedalam
#bentuk BeautifulSoup agar dapat diakses isi konten web nya
content = driver.page_source
soup = BeautifulSoup(content)
soup
```

```python
#setelah data htmlnya sudah didapatkan lalu dilakukan pencarian
#tag div beserta class nya kemudian diambil data gambar, produk
#nama produk, dan harga produk
#for melakukan perulangan untuk mengambil semua data yang ada class ta-productnya
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="card px-4"):
    print(link)
```

```python
#variabel dataImg mencari data dengan tag img dan class nya
#dan menampilkan dataImg nya berupa src yaitu link gambarnya
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="card px-4"):
    dataImg = link.find('img', class_="card-img-top")
    print(dataImg['src'])
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="card px-4"):
    dataImg = link.find('img', class_="card-img-top")
    
    deskripsi=link.find('span', class_="itemFacts")
    print(deskripsi.get_text())
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="itemPriceBox"):
    dataImg = link.find('img', class_="card-img-top")
    
    deskripsi = link.find('span', class_="itemFacts")
    
    price = link.find('span')
    print(price.get_text())
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="card px-4"):
    namaproduk = link.find('a', class_="itemName")
    print(namaproduk.get_text())
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
for link in soup.find_all('div', class_="card px-4"):
    dataImg = link.find('img', class_="card-img-top")
    
    deskripsi = link.find('span', class_="itemFacts")
    
    price = link.find('div', 'span', class_="itemPriceBox")
    
    namaproduk = link.find('a', class_="itemName")
#variabel product, prices, image ini mendefinisikan data yang
#sudah didapatkan dan memasukan ke dalam bentuk array
#yang sudah dibuat sebelumnya
    desk.append(deskripsi.get_text())
    prices.append(price.get_text())
    image.append(dataImg['src'])
    namaprod.append(namaproduk.get_text())
```

```python
#dari data diatas kemudian menyimpan nya ke dalam format .csv
df = pd.DataFrame({'image':image, 'namaproduk':namaprod, 'deskripsi':desk, 'price':prices})
df.to_csv('ikea7.csv', index=False, encoding='utf-8')
```

```python
#dari data diatas kemudian mentimpan nya ke dalam format .json
df = pd.DataFrame({'image':image, 'namaproduk':namaprod, 'deskripsi':desk, 'price':prices})
df.to_json('ikea7.json', orient = "table")
```
